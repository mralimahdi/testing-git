how to read and display single image

```python
import cv2
img=cv2.imread(r"C:\Users\DELL\Downloads\e1dd8989-67bc-4c27-86d4-060c418f3311.jpg")
cv2.imshow("window1",img)
#if the below lines are not mention then program will crash 
cv2.waitKey(0)
cv2.destroyAllWindows()
```

how to read and display multiple images but without using unique title for each window

```python
import cv2
img=cv2.imread(r"C:\Users\DELL\Downloads\e1dd8989-67bc-4c27-86d4-060c418f3311.jpg")
img1=cv2.imread(r"C:\Users\DELL\Downloads\b53f58ca-8bf1-4c8a-9d6d-5b261b6adf511-01.png")
cv2.imshow("window",img)
cv2.imshow("window",img1)
# if no unique name is given to the window then the last image will be displayed only
# if the below lines are not mention then program will crash 
cv2.waitKey(0)
cv2.destroyAllWindows()
```

how to read and display multiple images but by using unique title for each window

```python
import cv2
img=cv2.imread("C:/Users/DELL/Downloads/e1dd8989-67bc-4c27-86d4-060c418f3311.jpg")
img1=cv2.imread(r"C:\Users\DELL\Downloads\b53f58ca-8bf1-4c8a-9d6d-5b261b6adf511-01.png")
# typed the path manually with forward slash and it is working fine without writing r(raw string) before the path
cv2.imshow("window1",img)
cv2.imshow("window2",img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
