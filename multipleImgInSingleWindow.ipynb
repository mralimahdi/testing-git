how to resize image and display in vertical stack

```python
import cv2
import numpy as np

img=cv2.imread(r"C:\Users\DELL\Downloads\e1dd8989-67bc-4c27-86d4-060c418f3311.jpg")
re_img = cv2.resize(img, (400, 200))

v = np.vstack((re_img, re_img, re_img, re_img)) # number of rows should be same for vertical stacking

cv2.imshow("UIIT", v)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

how to resize image and display in horizontal stack

```python
import cv2
import numpy as np

img=cv2.imread(r"C:\Users\DELL\Downloads\e1dd8989-67bc-4c27-86d4-060c418f3311.jpg")
re_img = cv2.resize(img, (400, 200))

h = np.hstack((re_img, re_img, re_img, re_img)) # number of columns should be same for horizontal stacking

cv2.imshow("UIIT", h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

3x3 grid same images in each row

```python
import cv2
import numpy as np
img=cv2.imread(r"C:\Users\DELL\Downloads\e1dd8989-67bc-4c27-86d4-060c418f3311.jpg")
rimg=cv2.resize(img,(400,200))
row1=np.hstack((rimg,rimg,rimg))
grid=np.vstack((row1,row1,row1))
cv2.imshow("window",grid)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

3x3 grid with different images in each row

```python
import cv2
import numpy as np
img=cv2.imread(r"C:\Users\DELL\Downloads\e1dd8989-67bc-4c27-86d4-060c418f3311.jpg")
img2=cv2.imread(r"C:\Users\DELL\Downloads\b53f58ca-8bf1-4c8a-9d6d-5b261b6adf511-01.png")

rimg=cv2.resize(img,(400,200))
rimg2=cv2.resize(img2,(400,200))

row1=np.hstack((rimg,rimg2,rimg))
row2=np.hstack((rimg2,rimg,rimg))
row3=np.hstack((rimg,rimg2,rimg))
grid=np.vstack((row1,row2,row3))
cv2.imshow("window",grid)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
